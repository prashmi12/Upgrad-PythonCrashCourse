## Create a numpy array from the CSV dataset which should contain the top 10 states having the maximum number of zipcodes. Example, is StateA has 3 zipcodes, StateB has 8 zipcodes and StateC has 2 zipcodes, the output should be ["StateB", "StateA". "StateC"]

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/scpike/us-state-county-zip/master/geo-data.csv')
```

```python
df.head(10)
```

##### To check all the top countries having max zipcodes

```python
df.groupby('state')['zipcode'].size().sort_values(ascending=False).head(10)
```

```python
df.groupby(['state'])['zipcode'].count().sort_values(ascending = False).head(10)
```

```python
maxzipcode = df.groupby(['state'], as_index=False).sum()
```

```python
maxzipcode.rename(columns = {'state_fips':'total zipcodes'}, inplace = True)
```

```python
maxzipcode
```

```python
maxzipcode = maxzipcode.sort_values(by= 'total zipcodes', ascending= False)
```

```python
top10 = maxzipcode.head(10)['state'].values
```

```python
top10
```
